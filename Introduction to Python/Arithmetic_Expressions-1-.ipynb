Order of Operations Examples

```python
# easy example
example_1 = (14 - 7) / 7
print("Example 1 result is: " + str(example_1))

# challenging example
example_2 = ((((18 + 2) * 2) - 8) // 2) - 4 * 2

# Breakdown using order of operations to solve
# (18 + 2) = 20
# (20 * 2) = 40
# (40 - 8) = 32
# (32 // 2) = 16
# 4 * 2 = 8
# 16 - 8 = 8
print("Example 2 result is: " + str(example_2))

print(4 / 2)
```
